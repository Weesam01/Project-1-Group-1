```python
#Import libraries
import yfinance as yf
import pandas as pd
import seaborn as sns
%matplotlib inline
import hvplot.pandas
import matplotlib.pyplot as plt
```

```python
#Indicating the dates we want our date to be pulled from 
start = '2013-01-01'
end = '2023-10-04'
```

```python
#Indicating the Crypto we are going to be using for our Analysis 
cryptos = ['BTC-USD', 'ETH-USD', 'ADA-USD', 'XRP-USD']
```

```python
# Download Crypto Data into a Dataframe
df_cryptos = yf.download(cryptos, start=start, end=end)
```

```python
# Print the Dataframe results
df_cryptos
```

```python
df_cryptos = df_cryptos.dropna().copy() #Drop all null values
df_cryptos.head() # Print Results
```

```python
df_cryptos # Display Dataframe
```

```python
df_cryptos = df_cryptos.drop(columns=["Adj Close", "High", "Low", "Open"]) #  Drop specified Columns
df_cryptos.head()# Display Dataframe
```

```python
df_cryptos # Display Dataframe
```

```python
df_cryptos["Close"] = df_cryptos["Close"].pct_change(periods = 1)
df_cryptos.head() # Calculate Percent changes within the specified period 
```

```python
df_cryptos = df_cryptos.dropna().copy() # Drop Nulls
df_cryptos.head() #Display Dataframe
```

```python
df_cryptos["Close"].plot(figsize=(15,10),title="Volatility Chart "); # Creates a Line plot based on the Closing Data
```

```python
cryptos_cumulative_returns = (1 + df_cryptos["Close"]).cumprod() # Calculate the Cumulative Return of the close column
cryptos_cumulative_returns.head() # Display Datframe
```

```python
cryptos_cumulative_returns.plot(kind="line",figsize=(15,10),title="Collective Cumulative Returns of Portfolio Data "); # Plot Cumulative Return of Portfolio Data
```

```python
# Sort the DataFrame by the 'Volume' column in descending order
df_cryptos_volume = df_cryptos["Volume"].groupby(pd.Grouper(freq='1Y')).mean()

# Plot the sorted DataFrame (highest volume first)
df_cryptos_volume.plot(kind="bar", figsize=(15, 10), title="Annual Average Volume per Crypto")
plt.xlabel('Crypto')
plt.ylabel('Volume')
plt.show()
```

```python
# Plotting as a line chart
df_cryptos_volume.plot(kind="line", figsize=(15, 10), title="Annual Average Volume per Crypto")
plt.xlabel('Year')
plt.ylabel('Volume')
plt.show()
```

```python
crypto_std_deviation = df_cryptos["Close"].std() # Calculate Standard Deviation from the Close Column
crypto_std_deviation # Display Results
```

```python
crypto_std_deviation.plot(kind='barh'); #Create a Bar Plot based on the data above
```

```python
crypto_correlation = df_cryptos["Close"].corr() # Calculate Correlation from the Close Column
crypto_correlation # Display Results
```

```python
sns.heatmap(crypto_correlation, vmin=0, vmax=1, annot=True,); # Display Heatmap based above Data
```

```python
# Create a Weigted Portofolio Return. ADA,BTC,ETH,XRP
weights = [0.25, 0.25, 0.25, 0.25] # Weight average for each crypto
cryptos_cumulative_returns_portfolio = cryptos_cumulative_returns.dot(weights) # Assign a new variable for Weight calculation 
cryptos_cumulative_returns_portfolio.head() # Display Results
```

```python
initial_investment = 10000 # Initial Investment of 10,000
cryptos_cumulative_profit = initial_investment * cryptos_cumulative_returns_portfolio # Initial investment * the weight of the portfolio
cryptos_cumulative_profit.plot(figsize=(15, 10),title=" 25% Equal Weight: Return on Invesment of $10,000"); # Plot the Results
```

```python
# Define entry, target, and stop-loss prices for a single trade
entry_price = df_cryptos['Close']['BTC-USD']['2017-11-10']  # Start Date
target_price = entry_price * 1.10  # 10% profit target
stop_loss_price = entry_price * 0.95  # 5% stop-loss

# Calculate potential profit and potential loss 
risk_df=df_cryptos
risk_df['Potential_Profit'] = target_price - risk_df['Close', 'BTC-USD']
risk_df['Potential_Loss'] = risk_df['Close', 'BTC-USD'] - stop_loss_price

# Calculate the risk-reward ratio
risk_df['Risk_Reward_Ratio'] = risk_df['Potential_Profit'] / risk_df['Potential_Loss']

# Display the DataFrame with the risk-reward ratio
risk_df.head()
```

```python
# Drop Selected Columns
risk_df = risk_df.drop(columns=['Close', 'Volume'])
# Display the updated DataFrame
risk_df.head()
```

```python
# Create a line chart for 'Risk_Reward_Ratio' of BTC-USD over time
plt.figure(figsize=(12, 6))
plt.plot(risk_df.index, risk_df['Risk_Reward_Ratio'], label='BTC-USD', color='green')
plt.xlabel('BTC-USD')
plt.ylabel('Risk-Reward Ratio')
plt.title('BTC-USD Risk-Reward Ratio Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
