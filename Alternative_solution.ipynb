```python
import yfinance as yf
```

```python
start = '2013-01-01'
end = '2023-10-04'
```

```python
cryptos = ['BTC-USD', 'ETH-USD', 'ADA-USD', 'XRP-USD']
```

```python
df_cryptos = yf.download(cryptos, start=start, end=end)
```

```python
df_cryptos
```

```python
df_cryptos = df_cryptos.dropna().copy()
df_cryptos.head()
```

```python
df_cryptos
```

```python
df_cryptos = df_cryptos.drop(columns=["Adj Close", "High", "Low", "Open"])
df_cryptos.head()
```

```python
df_cryptos
```

```python
df_cryptos["Close"] = df_cryptos["Close"].pct_change(periods = 1)
```

```python
df_cryptos
```

```python
df_cryptos = df_cryptos.dropna().copy()
df_cryptos.head()
```

```python
df_cryptos["Close"].plot(figsize=(15,10))
```

```python
cryptos_cumulative_returns = (1 + df_cryptos["Close"]).cumprod()
cryptos_cumulative_returns.head()
```

```python
cryptos_cumulative_returns.plot(kind="line",figsize=(15,10))
```

```python
import pandas as pd
```

```python
df_cryptos_volume = df_cryptos["Volume"].groupby(pd.Grouper(freq='1Y')).mean()
```

```python
df_cryptos_volume.plot(kind="bar",figsize=(15,10), )
```

```python
crypto_std_deviation = df_cryptos["Close"].std()
crypto_std_deviation
```

```python
crypto_std_deviation.plot(kind='barh')
```

```python
crypto_correlation = df_cryptos["Close"].corr()
crypto_correlation
```

```python
import seaborn as sns
sns.heatmap(crypto_correlation, vmin=0, vmax=1, annot=True)
```

```python
weights = [0.25, 0.25, 0.25, 0.25]
cryptos_cumulative_returns_portfolio = cryptos_cumulative_returns.dot(weights)
cryptos_cumulative_returns_portfolio.head()
```

```python
initial_investment = 10000
cryptos_cumulative_profit = initial_investment * cryptos_cumulative_returns_portfolio
cryptos_cumulative_profit.plot(figsize=(15, 10))
```
